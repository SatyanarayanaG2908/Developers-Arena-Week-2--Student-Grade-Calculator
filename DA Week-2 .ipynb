```python
# Student Grade Calculator
# Week 2 Project - Decision Making & Loops
# Author: Your Name
# Date: December 25, 2025

"""
This program calculates student grades based on marks entered.
It demonstrates if-elif-else statements, while loops, functions, and input validation.
"""

def print_header():
    """Function to print a decorative header"""
    print("\n" + "=" * 60)
    print("ğŸ“ STUDENT GRADE CALCULATOR ğŸ“")
    print("=" * 60)

def print_footer():
    """Function to print a decorative footer"""
    print("=" * 60)
    print("Thank you for using the Student Grade Calculator! ğŸ“š")
    print("=" * 60 + "\n")

def get_grade(marks):
    """
    Function to determine grade based on marks
    Parameters: marks (int/float) - Student's marks
    Returns: tuple (grade, message)
    """
    if marks >= 90 and marks <= 100:
        return "A", "Outstanding! Excellent work! ğŸŒŸ"
    elif marks >= 80 and marks < 90:
        return "B", "Very Good! Keep it up! ğŸ‘"
    elif marks >= 70 and marks < 80:
        return "C", "Good! You're doing well! ğŸ˜Š"
    elif marks >= 60 and marks < 70:
        return "D", "Satisfactory! There's room for improvement! ğŸ’ª"
    elif marks >= 0 and marks < 60:
        return "F", "Keep trying! Don't give up! ğŸ¯"
    else:
        return None, None

def validate_marks(marks_str):
    """
    Function to validate if marks are in valid range
    Parameters: marks_str (string) - User input for marks
    Returns: tuple (is_valid, marks_value)
    """
    try:
        marks = float(marks_str)
        if marks >= 0 and marks <= 100:
            return True, marks
        else:
            return False, None
    except ValueError:
        return False, None

def display_result(name, marks, grade, message):
    """
    Function to display the final result in a formatted way
    Parameters: name, marks, grade, message
    """
    print("\n" + "=" * 60)
    print(f"ğŸ“Š RESULT FOR {name.upper()}")
    print("=" * 60)
    print(f"Student Name: {name}")
    print(f"Marks: {marks}/100")
    print(f"Grade: {grade}")
    print(f"Message: {message}")
    print("=" * 60)

def calculate_grade():
    """Main function to run the grade calculator"""
    print_header()

    # Get student name
    name = input("\nEnter student name: ").strip()

    # Validate name is not empty
    while name == "":
        print("âŒ Name cannot be empty! Please try again.")
        name = input("Enter student name: ").strip()

    # Get marks with validation using while loop
    marks = None
    while marks is None:
        marks_input = input(f"Enter marks for {name} (0-100): ").strip()
        is_valid, marks_value = validate_marks(marks_input)

        if is_valid:
            marks = marks_value
        else:
            print("âŒ Invalid input! Please enter a number between 0 and 100.")

    # Calculate grade
    grade, message = get_grade(marks)

    # Display result
    display_result(name, marks, grade, message)

def calculate_multiple_students():
    """Function to calculate grades for multiple students"""
    print_header()
    print("ğŸ“š Multiple Student Grade Calculator")
    print("=" * 60)

    # Get number of students
    num_students = None
    while num_students is None:
        try:
            num = int(input("\nHow many students do you want to grade? "))
            if num > 0:
                num_students = num
            else:
                print("âŒ Please enter a positive number!")
        except ValueError:
            print("âŒ Please enter a valid number!")

    # Store all results
    results = []

    # Use for loop to process multiple students
    for i in range(num_students):
        print(f"\n--- Student {i + 1} of {num_students} ---")

        # Get student name
        name = input("Enter student name: ").strip()
        while name == "":
            print("âŒ Name cannot be empty!")
            name = input("Enter student name: ").strip()

        # Get marks with validation
        marks = None
        while marks is None:
            marks_input = input(f"Enter marks for {name} (0-100): ").strip()
            is_valid, marks_value = validate_marks(marks_input)

            if is_valid:
                marks = marks_value
            else:
                print("âŒ Invalid input! Please enter a number between 0 and 100.")

        # Calculate grade
        grade, message = get_grade(marks)

        # Store result
        results.append({
            'name': name,
            'marks': marks,
            'grade': grade,
            'message': message
        })

    # Display all results
    print("\n" + "=" * 60)
    print("ğŸ“Š RESULTS SUMMARY")
    print("=" * 60)

    for i, result in enumerate(results, 1):
        print(f"\n{i}. {result['name']}")
        print(f"   Marks: {result['marks']}/100")
        print(f"   Grade: {result['grade']}")
        print(f"   {result['message']}")

    print("=" * 60)

def show_grading_system():
    """Function to display the grading system"""
    print("\n" + "=" * 60)
    print("ğŸ“– GRADING SYSTEM")
    print("=" * 60)
    print("Grade A: 90-100 marks - Outstanding! ğŸŒŸ")
    print("Grade B: 80-89 marks  - Very Good! ğŸ‘")
    print("Grade C: 70-79 marks  - Good! ğŸ˜Š")
    print("Grade D: 60-69 marks  - Satisfactory! ğŸ’ª")
    print("Grade F: 0-59 marks   - Keep trying! ğŸ¯")
    print("=" * 60)

def main():
    """Main program function with menu"""
    while True:
        print("\n" + "=" * 60)
        print("ğŸ“ STUDENT GRADE CALCULATOR - MAIN MENU")
        print("=" * 60)
        print("1. Calculate grade for one student")
        print("2. Calculate grades for multiple students")
        print("3. View grading system")
        print("4. Exit")
        print("=" * 60)

        choice = input("\nEnter your choice (1-4): ").strip()

        if choice == "1":
            calculate_grade()
        elif choice == "2":
            calculate_multiple_students()
        elif choice == "3":
            show_grading_system()
        elif choice == "4":
            print("\n" + "=" * 60)
            print("ğŸ‘‹ Thank you for using the Grade Calculator!")
            print("Keep learning and growing! ğŸ“šâœ¨")
            print("=" * 60 + "\n")
            break
        else:
            print("\nâŒ Invalid choice! Please enter a number between 1 and 4.")

# Run the program
if __name__ == "__main__":
    main()
```
